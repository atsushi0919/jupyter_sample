# グラフも描ける！

でも、`github`上ではJavaScriptは動作せず、グラフが表示されないので...

`nbviewer` というビューワーを通すと、グラフも表示される！

````ruby
require 'rbplotly'

trace1 = { x: [1,2,3,4,5,6], y: [100,150,300,350,400,500] }
trace2 = { x: [1,2,3,4,5,6], y: [600,550,600,250,180,100] }
traces = [trace1, trace2]
pl = Plotly::Plot.new(data: traces)
pl.show
````
